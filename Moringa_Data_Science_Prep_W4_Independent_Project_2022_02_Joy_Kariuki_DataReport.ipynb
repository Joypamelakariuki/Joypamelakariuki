## Loading the csv dataset to be analysed

```python
cars_dataset = pd.read_csv('http://bit.ly/autolib_dataset')
#cars_dataset.head()
```

## Changing the dataset into a csv file

```python
cars_data_csv = cars_dataset.to_csv('cars_data_csv',encoding='utf-8',index=False)
cars_data = pd.read_csv('cars_data_csv')
cars_data.head(10)
```

-After loading the data, it is important to first understand how the data is. The colums names and the data, and how they can be used together.

```python
cars_data.info()
```

## The  next step is to clean the data. This is done by dropping the columns that are not important or won't be  used. Among those that will be used include: Address, Bluecars, City, Month, Postal code, Station, Utilib, utilib 1.4, Hour.

```python
cars_data_clean = cars_data.drop(['Cars','Charge Slots','Charging Status','Rental status','Slots','year','minute','Displayed comment','Geo point','Scheduled at'],axis=1)
cars_data_clean.head(5)
```

The next step is to drop the duplicated rows for easier analysis

```python
cars_data_clean = cars_data_clean.drop_duplicates()
cars_data_clean.head(5)
```

## Analysis

Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
##Changing the colums names for easier use
cars_data_clean.columns = ['Address', 'Bluecar_counter', 'Utilib_counter', 'Utilib1_counter', 'City', 'ID','Kind', 'Postal_code','public_name','station_type','Status','subscription_status','month','day','hour']
cars_data_clean.head(5)
```

```python
#Picking the data for Paris city only
cars_data_Paris = cars_data_clean[cars_data_clean.City == 'Paris']
cars_data_Paris.head(20)
```

```python
#Create another column for the usage of the blue car, if it was returned, its represented by a negative, if its picked, positive and 0 for none happening
cars_data_Paris['Bluecar_usage'] = cars_data_Paris.Bluecar_counter.diff()
cars_data_Paris
```

# Picking the hour popular for picking up in Paris

```python
#Picking the hour popular for picking up shared electricc car(Bluecar). We are picking a negative number since it represents cars being picked.
cars_data_Paris[(cars_data_Paris.City == 'Paris') & (cars_data_Paris.Bluecar_usage > 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(5)
```

From the data, the hour 10:00 is the most popular which is 10:00 am. for picking the cars.

# Most popular hour for returning the cars


```python
#In  this case, we are not only considering Paris
cars_data_clean['Bluecar_usage'] = cars_data_clean.Bluecar_counter.diff()
cars_data_clean
```

```python
cars_data_clean[(cars_data_clean.Bluecar_usage < 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(5)
```

21hours was the most popular hour for returning the cars which is 9:00pm.

# What station is the most popular overall
