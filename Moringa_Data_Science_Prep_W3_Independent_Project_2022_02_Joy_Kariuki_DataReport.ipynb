**The next step is to clean the data and Analyse it**

````python
#To clean the data, we have to combine the telecom datasets and drop the unecessary columns that we don't need.
````

````python
#Combining the telecom datasets using concatenate
telecom = pd.concat([teldat_1,teldat_2,teldat_3], axis=0)
telecom.info()
````

````python
##The next step is to drop the columns that we don't need.
##Drop the country columns from the dataset
telecom_data = telecom.drop(['Country_A', 'Country_B'], axis=1)
##The Cell_ID column will be used when concatenating with the geo dataset
````

````python
#To be able to calculate the time on the Date Time column, we have to convert the data to a format easy to work with
telecom_data['Date_Time'] = pd.to_datetime(telecom['Date_Time'])
````

````python
telecom_data
````

````python
##The rows in the dataset are many, dropping duplicated would help reduce them and also increase the quality of the data
telecom_data = telecom_data.drop_duplicates()
````

````python
##Viewing the data after dropping the duplicates
telecom_data
````

````python
##The next step will be to merge the telecom_data dataset with the cell_geo dataset for further cleaning 
#We will merge them on the Cell ID for the different areas
telecom_geo_data = pd.merge(telecom_data,geo, on=['Cell_ID'])
````

````python
##Viewing the data
telecom_geo_data.info()
````

### **Data Analysis/Exploration**

````python
##To explore the data, we have to check which product was hevily used and in which location.
#This will help the company know which are areas use the different products more heavily.
#The next step will be to determine the cities that mostly used the products.
#Determining according to the time, which ones used for longer time
````

````python
##Determining the total usage of the products for the entire dataset
telecom_data.groupby('Product')['Value'].sum()
````

From the output given, voice was leading, followed by data and finally sms.

````python
##Determining the cities that had the highest values for the products usage
telecom_data.groupby('Cell_ID').count().nlargest(5,'Value')
````
